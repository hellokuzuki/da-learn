## Datetime handling

```python
# parse_datas
# If True -> try parsing the index.
# If [1, 2, 3] -> try parsing columns 1, 2, 3 each as a separate date column.
# If [[1, 3]] -> combine columns 1 and 3 and parse as a single date column.\
# If {'foo': [1, 3]} -> parse columns 1, 3 as date and call result ‘foo’. A fast-path exists for iso8601-formatted dates.
read_datetype_o = pd.read_csv('./read_csv_example.csv', skip_blank_lines=True)
print(read_datetype_o['last_modified_on'].dtype)
read_datetype_d = pd.read_csv('./read_csv_example.csv', skip_blank_lines=True, parse_dates=['last_modified_on'])
print(read_datetype_d['last_modified_on'].dtype)
read_datetype_combine = pd.read_csv('./read_csv_example.csv', skip_blank_lines=True, parse_dates={'foo': [8, 10]})
read_datetype_combine.head(5)
read_datetype_combine_keep_col = pd.read_csv('./read_csv_example.csv', skip_blank_lines=True, parse_dates={'foo': [8, 10]}, keep_date_col=True)
read_datetype_combine_keep_col.head(5)
```

```python
# date_parser
from datetime import datetime
read_date_parser = pd.read_csv('./read_csv_example.zip', skip_blank_lines=True, parse_dates=['created_on'], date_parser= lambda x : datetime.strptime(x, '%d/%m/%Y %H:%M'), compression='zip')
read_date_parser['created_on'].dtype
```

## from Dealing with Unicode data

```python
from io import BytesIO
data = b"word,length\n" b"Tr\xc3\xa4umen,7\n" b"Gr\xc3\xbc\xc3\x9fe,5"
data = data.decode('utf8').encode('latin-1')
df = pd.read_csv(BytesIO(data), encoding="latin-1")
print(df["word"][1])
```
